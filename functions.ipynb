Problem Statement 1:Smart Parking Lot Management System

```python
def parking_system(capacity, logs):
    
    parked = 0
    peak = 0
    # Check each vehicle log
    for log in logs:
        if log == "IN":
            parked = parked + 1
        elif log == "OUT":
            parked = parked - 1
        # Track peak usage
        if parked > peak:
            peak = parked
    print("Currently Parked Vehicles:", parked)
    # Check parking status
    if parked > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")
# Sample Input
capacity = 50
logs = ["IN","IN","IN","OUT","IN","IN","OUT"]
parking_system(capacity, logs)
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):
     # Basic time calculation
    time = distance * 5
# Traffic condition
    if traffic == "High":
        time = time + 15
    elif traffic == "Medium":
        time = time + 10
    else:
        time = time + 5
# Weather condition
    if weather == "Rainy":
        time = time + 10
    elif weather == "Storm":
        time = time + 20
    print("Estimated Delivery Time:", time, "minutes")
# Sample Input
delivery_time(8,"High","Rainy")
```

Problem Statement 3:Movie Theatre Seat Occupancy
Analyzer

```python
def theatre_seats(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100
    print("Occupancy:", int(percentage),"%")
    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
# Sample Input
booked = [1]*150
theatre_seats(200,booked)
```

Problem Statement 4:Cloud Server Load Classification System

```python
def server_load(cpu_list):
    total = 0
# Add all values
    for value in cpu_list:
        total = total + value
    average = total / len(cpu_list)
    print("Average CPU Load:", int(average),"%")
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
# Sample Input
cpu = [45,60,70,85,90]
server_load(cpu)
```

Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def classroom_usage(resources):
    overused = []
    # Check usage
    for item in resources:
        if resources[item] > 8:
            overused.append(item)
    print("Overused Resources:",",".join(overused))
    if len(overused) > 0:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")
# Sample Input
data = {
"Projector":6,
"AC":9,
"Lights":4
}
classroom_usage(data)
```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        print("Confirmed Registrations:",confirmed)
        print("Waitlisted Users:",waitlist)
        print("Registration Status: Closed")
    else:
        print("Confirmed Registrations:",registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
# Sample Input
event_registration(100,105)
```
